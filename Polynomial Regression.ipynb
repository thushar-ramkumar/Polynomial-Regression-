```python
import pandas as pd
data = pd.read_csv("Book2.csv",na_values=-200)
```

```python
data.columns
data.isnull().sum()
data.info()
data.shape
```

```python
data.dropna(thresh=7,axis=0,inplace=True)
data.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1, inplace=True)
data.drop('NMHC(GT)', axis=1, inplace=True)
```

```python
data['Date']=pd.to_datetime(data.Date, format='%m/%d/%Y') 
data.set_index('Date',inplace=True)
```

```python
data['Month']=data.index.month 
data.reset_index(inplace=True)
```

```python
data['Hour']=data['Time'].apply(lambda x: int(x.split(':')[0]))
data
```

```python
data['CO(GT)']=data['CO(GT)'].fillna(data.groupby(['Month','Hour'])['CO(GT)'].transform('mean'))
data['NOx(GT)']=data['NOx(GT)'].fillna(data.groupby(['Month','Hour'])['NOx(GT)'].transform('mean'))
data['NO2(GT)']=data['NO2(GT)'].fillna(data.groupby(['Month','Hour'])['NO2(GT)'].transform('mean'))
```

```python
data.isnull().sum()
```

```python
data['CO(GT)']=data['CO(GT)'].fillna(data.groupby(['Hour'])['CO(GT)'].transform('mean'))
data['NOx(GT)']=data['NOx(GT)'].fillna(data.groupby(['Hour'])['NOx(GT)'].transform('mean'))
data['NO2(GT)']=data['NO2(GT)'].fillna(data.groupby(['Hour'])['NO2(GT)'].transform('mean'))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = data[:].drop(['RH', 'Date', 'Time'] ,axis=1)
y = data['RH']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

y_pred = lr.predict(X_test)
rmse = round(np.sqrt(mean_squared_error(y_test,y_pred)),2)

plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Linear Regression')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'fit_intercept':[True, False], 'normalize':[True, False]}

GS = GridSearchCV(lr, parameters,cv=5)
GS.fit(X_train,y_train)

GS.best_params_
GS.best_score_
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

polyfeatures = PolynomialFeatures()
X_poly = polyfeatures.fit_transform(X)
X_poly = X_poly[:,1:]

X_train, X_test, y_train, y_test = train_test_split(X_poly,y,random_state=0)

polyreg = LinearRegression()
polyreg.fit(X_train, y_train)
polyreg.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

y_pred = polyreg.predict(X_test)
rmse = round(np.sqrt(mean_squared_error(y_test,y_pred)),2)

plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Polynomial Regression')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'fit_intercept':[True, False], 'normalize':[True, False]}

GS = GridSearchCV(polyreg, parameters,cv=5)
GS.fit(X_train,y_train)

GS.best_params_
GS.best_score_
```
